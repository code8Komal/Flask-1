### Q1. What is Flask Framework? What are the advantages of Flask Framework?

# Ans 
Flask is a web framework for building web applications using the Python programming language
Some advantages of using Flask include:
1. Lightweight and flexible
2. Easy to learn
3. Extensible
4. Large community
5. Compatibility with other technologies

### Q3. What is App routing in Flask? Why do we use app routes?







#### Ans
App routing refers to the process of mapping URLs (Uniform Resource Locators) to specific functions or views within a Flask application.
@app.route() decorator to define routes. This decorator is applied to a function, and it associates the function with a specific URL or URL pattern
